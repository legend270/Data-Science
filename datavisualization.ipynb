```python
import matplotlib.pyplot as plt
```

```python
x_values = ['Python','C++','C','Java']
y_values = [25,100,34,56]
```

```python
plt.plot(x_values, y_values, color='g',marker='h',linestyle='-')
plt.xlabel('Programming language')
plt.ylabel('Votes')
plt.title('vote casting')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(x_values, y_values, color='r',edgecolor='g')
plt.xlabel('Programming language')
plt.ylabel('Votes')
plt.title('vote casting')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.pie(y_values, labels=x_values, autopct='%1.0f%%', explode=[0.05,0.05,0.05,0.15], shadow=False)
plt.title('Dataset')
plt.savefig("plot.png",dpi=500)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('FINE GRAINED SENTIMENTS.csv')
df
```

```python
value_count = df.value_counts('FINE-GRAINED')
value_count
```

```python
x_values = value_count.index
y_values= value_count.values
```

```python
plt.bar(x_values, y_values, color='r',edgecolor='g')
plt.xlabel('Programming language')
plt.ylabel('Votes')
plt.title('vote casting')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.pie(y_values, labels=x_values)
plt.title('Dataset')
plt.show()
```

```python
plt.plot(x_values, y_values, color='g',marker='h',linestyle='-')
plt.xlabel('Programming language')
plt.ylabel('Votes')
plt.title('vote casting')
plt.xticks(rotation=45)
plt.show()
```
